```
import pandas as pd

path = 'https://raw.githubusercontent.com/commit-live-students/linear_regression_project/master/data/house_prices_multivariate.csv'
def load_data(path):
    df=pd.read_csv(path)
    return df
load_data(path)
```

```
df = load_data('https://raw.githubusercontent.com/commit-live-students/linear_regression_project/master/data/house_prices_multivariate.csv')
def data_splitter(df):
    X = df.iloc[:,:-1]
    y = df.SalePrice
    return X,y
     
data_splitter(df)
```
